# Upload The Data That You Wnat To Work With It 

```python
import pandas as pd 
df = pd.read_csv(r"C:\Users\LENOVO\Downloads\pandas_exercise_big_dataset.csv")
print(df.head())
```

# Find the average score of all people

```python
average_score = df["score"].mean()
print(average_score)
```

# Find the oldest person’s name and age

```python
oldest_row = df.loc[df["age"].idxmax()]

print(oldest_row["name"], oldest_row["age"])
```

# Determine how many people live in each city

```python
city_counts = df["city"].value_counts()
print(city_counts.to_string())
```

# Filter the dataset to only include rows where score ≥ 90

```python
high_scores = df[df["score"] >= 90]
high_scores
```

# Add a new column called “age_group”:

# "young" if age < 25

# "adult" if 25 ≤ age < 35

# "senior" if age ≥ 35

```python
def age_group(age):
    if age < 25:
        return "young"
    elif age < 35:
        return "adult"
    else:
        return "senior"

df["age_group"] = df["age"].apply(age_group)
df.head()
```
