Level 1

Load the CSV file (ignore the header row)

What is the shape of the array

How many rows and columns are there

What’s the data type of the array


```python
import numpy as np
data = np.loadtxt(r"C:\Users\LENOVO\Downloads\sales_data.csv", delimiter=',', skiprows=1, usecols=(1,2,3))
print(data)
```

```python
array = data
print(array.shape)
```

```python
rows, cols = array.shape
print("Rows:", rows)
print("Columns:", cols)
```

Level 2 : Basic Calculation

The total sales for each product (sum across all months).
# total per product

The average sales per product.
# average per product

The maximum and minimum sales for each product.
# min per product
# max per product 

```python
print(np.sum(array, axis=0))
print(np.mean(array, axis=0))
print(np.min(array, axis=0)) 
print(np.max(array, axis=0))
```

Level 4 — Simple Transformation

Suppose the company increases all sales by 10% — create a new array with updated numbers.

```python
increased_sales = array * 1.10
print(increased_sales)
```

Level 5 — Bonus Challenge

Add a new column with the total sales per month and print the result

```python
total_per_month = np.sum(array, axis=1)
print(total_per_month)
```

```python
new_data = np.column_stack((array, total_per_month))
print(new_data)
```
