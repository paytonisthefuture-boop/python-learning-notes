# Part D — GroupBy & Aggregation

10. Group by Pclass and compute average Age and Fare.

11. Group by Sex and calculate passenger count and survival rate.

12. Group by Pclass and Sex and compute average Fare.

13. For each Embarked port, find the count of survivors

```python
df.groupby(['pclass', 'fare'])['age'].mean()
```

```python
df.groupby(['name', 'sex'])['survived'].count()
```

# Part E — Combined Operations

14. Among survivors, compute average Fare by Pclass and sort descending.

15. For each Pclass, find the passenger with the highest Fare.

```python
df[df['survived'] == 1].groupby('pclass')['fare'].mean().sort_values(ascending=False)
```
