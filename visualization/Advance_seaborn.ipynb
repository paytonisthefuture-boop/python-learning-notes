```python
import seaborn as sns 
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size="size")
```

```python
sns.histplot(tips['tip'], kde=True, bins=15)
```

```python
sns.displot(tips['tip'], kde=True, bins=20) 
```

```python
sns.barplot(x="sex", y="tip", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex")
```

```python
sns.stripplot(x="day", y="tip", data=tips, hue="sex", dodge=True)
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="kde", fill=True)
```

```python
sns.pairplot(titanic.select_dtypes(['number']), hue="pclass")
```

```python
iris.drop("species", axis=1).corr()
sns.heatmap(iris.drop("species", axis=1).corr(), annot=True)
```

```python
sns.clustermap(iris.drop("species", axis=1))
```
