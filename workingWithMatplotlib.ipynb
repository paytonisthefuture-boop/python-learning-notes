# Labels

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([100, 50, 44, 12])
y3 = np.array([13, 15, 20, 30])

plt.title("class Size", fontsize=25, 
          family="Arial", 
          fontweight="bold", 
          color="#0A44AF")

plt.xlabel("Year", fontsize=25, 
          family="Arial", 
          fontweight="bold", 
          color="#9BDE0A")

plt.ylabel("Students", fontsize=25, 
          family="Arial", 
          fontweight="bold", 
          color="#9BDE0A")
plt.tick_params(axis="both", 
                colors="#9BDE0A")

line_style = dict(marker=".", 
         markersize=30, 
         markerfacecolor="#11C3CA", 
         markeredgecolor="#11C3CA",
         linestyle="solid", 
         linewidth=4)


plt.plot(x, y1, color="#11C3CA",  **line_style)
plt.plot(x, y2, color="#CD0909",  **line_style) 
plt.plot(x, y3,color="#066707",  **line_style)
plt.xticks(x)

plt.show()
```

# Grid() = Helps Make Plots Easier To Read By Adding Reference Lines. 

```python
x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]
plt.grid(axis="y", 
         linewidth=2, 
         color="lightgray", 
         linestyle="dotted")
plt.plot(x, y)
plt.show()
```

# Bar Chart = Compare Categories Of Data By Representing Each Category With A Bar 

```python
categories = np.array(["Grains", "Fruits", "Vegetables", "protein", "Dairy", "Sweets"])
values = np.array([4, 3, 2, 5, 3, 1])


plt.bar(categories, values, color="skyblue")
# plt.bar(categories, values, color="skyblue")


plt.title("Daily consmption")
plt.xlabel("Food")
plt.ylabel("Quantity")


plt.show()
```

# Bar Chart = Circular Chart Divided Into Slices To Show Percentages Of Total.
#            Good For Visualizing Distribution Among Cetegories. 

```python
categories = ["Freshmen", "Sophommores", "Juniors", "Seniors"]
values = np.array([300, 250, 275, 225])

colors = ["red", "yellow", "blue", "green"]

plt.pie(values, labels=categories, 
        autopct="%1.1f%%", 
        colors=colors, 
        explode=[0, 0, 0, 0.1], 
        shadow=True, 
        startangle=90)

plt.title("Bilals College")
plt.show()
```

# Scatter Graph = Shows The Relationship Between Two Variables 
#                Helps To Identify A Correlation (+, -, None)
#                Example: Study Hours Vs. Test Scores

```python
x1 = np.array([0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8]) # Hours Study
y1 = np.array([55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87]) # Grades

x2 = np.array([0, 1, 2, 2, 3, 4, 5, 6, 7, 8, 8]) # Hours Study
y2 = np.array([55, 58, 65, 70, 72, 78, 83, 88, 92, 95, 97]) # Grades




plt.scatter(x1, y1, color="red", 
             alpha=0.5, 
             s = 100, 
             label="Class A") 


plt.scatter(x2, y2, color="skyblue", 
             alpha=0.5, 
             s = 100, 
             label="Class B")


plt.title("Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Grade") 
 
plt.legend()
 
plt.show()
```

# Histogram = A Visual Representation Of The Distribution Of Quamtitative Data .
#             They Group Values Into Bins (intervals)
#             And Counts How Many Fall In Each Ramge.

```python
scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores, 0, 100)

plt.hist(scores, bins=10, 
         color="lightgreen", 
         edgecolor="black",
    )
plt.title("Exam Scores")
plt.xlabel("Score")
plt.ylabel("# of studennts")



plt.show()
```

# Figure = The Entire Canvas
# Ax = A Single Plot (Subplot)

```python
figure, axes = plt.subplots(2, 2) 

x = np.array([1, 2, 3, 4, 5])

# Access the Axes object first, then call methods
axes[0, 0].plot(x, x * 2, color="red")
axes[0, 0].set_title("x*2")

axes[0, 1].plot(x, x**2, color="blue")
axes[0, 1].set_title("x**2")

axes[1, 0].plot(x, x**3, color="green")
axes[1, 0].set_title("x**3") # Corrected title typo

axes[1, 1].plot(x, x**4, color="purple")
axes[1, 1].set_title("x**4")

plt.tight_layout()
plt.show()
```
